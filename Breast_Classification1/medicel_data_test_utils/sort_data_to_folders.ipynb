```python
import os
import pandas as pd
import shutil
```

```python

labels_file = '../../data/breast/train_phase2/train_phase2/labels.csv'
root = '../../'
img_root = '../../data/breast/train_phase2/train_phase2/grey_original'

df = pd.read_csv(labels_file)
root = os.path.join(root,'split_data')
os.makedirs(root, exist_ok=True)

cancer_path = os.path.join(root,'cancer')
benign_path = os.path.join(root,'benign')

os.makedirs(cancer_path,exist_ok=True)
os.makedirs(benign_path,exist_ok=True)
```

```python
df
```

```python
image_names = df['PatientID']+"_"+df['StudyUID']+"_"+df['View']+"_"+df['Slice'].apply(str)+"_grey.png"
```

```python
image_names
```

```python
for img,lbl in zip(image_names,df['Class']):
    if lbl == 'cancer':
        lbl = 1
    elif lbl == 'benign':
        lbl = 0
    else:
        print('no lable found')
        pass
    
    img_path = os.path.join(img_root, img)
    if lbl:
        save_path = os.path.join(cancer_path,img)
    else:
        save_path = os.path.join(benign_path,img)
        
    print(img)
    
    shutil.copy(img_path, save_path)
```
